```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
```

```python
#Construct a simple timeseries
dates = pd.date_range('1/1/2000', periods=8)
df = pd.DataFrame(np.random.randn(8, 4),index=dates, columns=['A', 'B', 'C', 'D'])
df
```

```python
#准确定位到某个项，比如2010-01-06的A
df['A'][dates[5]]
```

```python
#You can pass a list of columns to [] to select columns in that order. 
#If a column is not contained in the DataFrame, an exception will be raised. Multiple columns can also be set in this manner:
df[['B', 'A']] = df[['A', 'B']]
df
```

### Selection by callable

```python
df1 = pd.DataFrame(np.random.randn(6, 4),index=list('abcdef'),columns=list('ABCD'))
df1
```

```python
df1.loc[lambda df: df['A'] > 0, :]
```

```python
df1.loc[:, lambda df: ['A', 'B']]
```

```python
df1.loc[:,['A','B']]   #和上面结果一样
```

```python
df1.iloc[:, lambda df: [0, 1]]
```

```python
df1[lambda df: df.columns[0]]
```

## Example

```python
df1=pd.DataFrame({'X':range(5),'Y':range(5),'S':list("aaabb"),'Z':[1,1,2,2,2]})
df1
```

```python
### 指定单个列为索引
df1.set_index('S')
```

```python
### 同时指定多个列为索引
df1.set_index(['S','Z'])
```

```python
#保留作为索引的列
df1.set_index(['S','Z'],drop=False)    
```

```python
## 索引转换为列
df2=df1.set_index(['S','Z'])
df2
```

```python
### 单个索引转换为列
df2.reset_index('Z')
```

```python
### 多个索引转换为列
df2.reset_index()
```

```python
### 直接删除对指定索引
df2.reset_index('Z',drop=True)
```

```python
### 直接对原DataFrame对象修改
df2.reset_index(inplace=True)
df2
```

```python
import pandas as pd
import numpy as np
```

```python
data_1=pd.read_csv('D://data/titanic_test.csv')
data_1.head()
```

## value_counts()

```python
data_1['Age'].value_counts()   #默认降序
```

```python
data_1['Age'].value_counts(ascending=True)   #升序
```

```python
data_1['Age'].value_counts(normalize=True)   #统计下概率分布
```

```python
type(data_1['Age'].value_counts(normalize=True))  #可以看出这是一个series
```

```python
data_1['Age'].value_counts(ascending=True)/sum(data_1['Age'].value_counts())
```

空值是默认剔除掉的，value_counts()返回的结果是一个Series数组，可以跟别的数组进行运算。

value_count()跟透视表里（pandas或者excel）的计数很相似，都是返回一组唯一值，并进行计数，这样能快速找出重复出现的值。

还有value_counts()函数是针对Series的，不是针对DataFrame的，所以只能是单列。

## descirbe()

```python
#describe函数总结数据集分布的中心趋势，分散和形状，不包括NaN值。
data_1.describe()
```

```python
#include包含all、[np.number]和[np.object]三个值
#describe属性可以对数值型变量（include=['number']）和离散型变量（include=['object']）进行描述性统计：
data_1.describe(include='all')
```

```python
data_1.describe(include=[np.number])
```

```python
data_1.describe(include=[np.object])
```
